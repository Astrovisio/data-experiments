```python
import pynbody
import matplotlib.pyplot as plt
import pylab
```

```python
f = pynbody.snapshot.ascii.AsciiSnap(".data/SERRA/zinnia_test_data/serra05_s46_h6043_test_data.ascii")
#pos_x   pos_y   pos_z   size    

# Compute the mass
mass = f["rho"] * f["size"]**3

# Create a SimArray for mass with appropriate units
f['mass'] = pynbody.array.SimArray(mass, units="Msol")

f.physical_units()
```
